#Load and Inspect the Data

```python
import pandas as pd

# Load the dataset
heartrate_df = pd.read_csv('heartrate_seconds_merged.csv')

# Display the first few rows and column names
print(heartrate_df.head())
print(heartrate_df.columns)
```

#Data Cleaning

```python
import pandas as pd

# Load the dataset
heartrate_df = pd.read_csv('heartrate_seconds_merged.csv')

# Display initial missing values and data types
print("Initial Missing Values:")
print(heartrate_df.isna().sum())
print("\nInitial Data Types:")
print(heartrate_df.dtypes)

# Drop rows with missing values
heartrate_df = heartrate_df.dropna()

# Convert 'Time' to datetime format if it exists
if 'Time' in heartrate_df.columns:
    heartrate_df['Time'] = pd.to_datetime(heartrate_df['Time'], errors='coerce')

# Verify the cleaning results
print("\nAfter Cleaning Missing Values:")
print(heartrate_df.isna().sum())
print("\nData Types After Cleaning:")
print(heartrate_df.dtypes)
```

#Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Distribution of Heart Rate
plt.figure(figsize=(10, 6))
sns.histplot(heartrate_df['Value'], kde=True)  # Using 'Value' for heart rate
plt.title('Distribution of Heart Rate')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')
plt.show()

# Plot Heart Rate Over Time
if 'Time' in heartrate_df.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(heartrate_df['Time'], heartrate_df['Value'], marker='o')
    plt.title('Heart Rate Over Time')
    plt.xlabel('Time')
    plt.ylabel('Heart Rate')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Optional: Plot Heart Rate by Hour of Day
if 'Time' in heartrate_df.columns:
    heartrate_df['Hour'] = heartrate_df['Time'].dt.hour
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Hour', y='Value', data=heartrate_df)  # Using 'Value' for heart rate
    plt.title('Heart Rate by Hour of Day')
    plt.xlabel('Hour of Day')
    plt.ylabel('Heart Rate')
    plt.show()
```

# Save the cleaned data

```python
import pandas as pd
from google.colab import files  # Import the files module

# Load the dataset
heartrate_df = pd.read_csv('heartrate_seconds_merged.csv')

# Data Cleaning
heartrate_df = heartrate_df.dropna()
if 'Time' in heartrate_df.columns:
    heartrate_df['Time'] = pd.to_datetime(heartrate_df['Time'], errors='coerce')

# Save the cleaned data to a CSV file
cleaned_file_path = '/content/clean_heartrate_data.csv'
heartrate_df.to_csv(cleaned_file_path, index=False)

print(f"Cleaned data saved to {cleaned_file_path}")

# Download the cleaned CSV file
files.download(cleaned_file_path)
```
