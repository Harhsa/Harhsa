#Set Up the Environment

```python
import sqlite3
import pandas as pd

# Connect to the database
con = sqlite3.connect('/sql-murder-mystery.db')

# Confirm connection
print(con)
```

#Retrieve Crime Scene Report

```python
query_1 = '''
SELECT *
FROM crime_scene_report
WHERE city = "SQL City"
ORDER BY date;
'''

#seting the dataframe width to max
pd.set_option('display.max_colwidth', None)

#running our query
pd.read_sql_query(query_1, con)
```

#Retrieve Witnesses' Personal Details

```python
query_2 = '''
SELECT *
FROM Person
WHERE address_street_name = "Northwestern Dr"
ORDER BY address_number DESC;
'''

witnesses = pd.read_sql_query(query_2, con)
witnesses
```

#View Witness Interviews

```python
query_3 = '''
SELECT *
FROM interview
WHERE person_id = 14887 OR person_id = 16371;
'''

witness_interviews = pd.read_sql_query(query_3, con)
witness_interviews
```

#Investigate the Gym Database

```python
query_4 = '''
SELECT *
FROM get_fit_now_check_in
WHERE membership_id LIKE '%48Z%' AND check_in_date = 20180109
ORDER BY check_in_date;
'''

gym_check_ins = pd.read_sql_query(query_4, con)
gym_check_ins
```

#Examine Car Details

```python
query_5 = '''
SELECT *
FROM drivers_license
WHERE plate_number LIKE "%H42W%";
'''

car_details = pd.read_sql_query(query_5, con)
car_details
```

#Check Personal Details

```python
query_6 = '''
SELECT *
FROM Person
WHERE license_id IN ("423327", "664760");
'''

personal_details = pd.read_sql_query(query_6, con)
personal_details
```

#Confirm Gym Membership Status

```python
query_7 = '''
SELECT *
FROM get_fit_now_member
WHERE person_id IN ("51739", "67318");
'''

gym_membership = pd.read_sql_query(query_7, con)
gym_membership
```

#Analyze Facebook Event Attendance

```python
query_8 = '''
SELECT person_id, count(*), event_name
FROM facebook_event_checkin
GROUP BY person_id
HAVING count(*) = 3 AND event_name = "SQL Symphony Concert" AND date LIKE '%201712%';
'''

event_checkins = pd.read_sql_query(query_8, con)
event_checkins
```

#Draw Conclusions with Queries
