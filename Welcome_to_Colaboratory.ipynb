#Load the Dataset

```
import pandas as pd

# Load the dataset
df = pd.read_csv('dailyActivity_merged.csv')
```

#Data Cleaning

```
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'], format='%m/%d/%Y')
```

Check for Duplicates and Remove Them

```
df.drop_duplicates(inplace=True)
```

Check for Missing Values

```
missing_values = df.isnull().sum()
print(missing_values)
```

#Exploratory Data Analysis

Plotting Distributions and Correlations:

```
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['TotalSteps'], kde=True)
plt.show()

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

#Data Transformation

```
df['ActiveMinutes'] = df['VeryActiveMinutes'] + df['FairlyActiveMinutes'] + df['LightlyActiveMinutes']
```

#Data Export for Tableau

```
df.to_csv('cleaned_dailyActivity.csv', index=False)
```
